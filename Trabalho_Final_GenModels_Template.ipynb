````python
# --- 2. INGESTAO E TEXT SPLITTER ---
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# API Key fornecida eu estou usando minha chave de API, talvez eu tenha excluido elas antes de publicar no portal, altere 
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY

pdf_path = arquivo_pdf
try:
    loader = PyPDFLoader(pdf_path)
    docs = loader.load()
    print(f"Total de paginas carregadas: {len(docs)}")

    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    splits = text_splitter.split_documents(docs)
    print(f"Total de de chunks (pedaços) extraídos para o banco: {len(splits)}")
    
    # Demonstando um trecho isolado
    if len(splits) > 0:
        print(f"\nExemplo de leitura (1º chunk): {splits[0].page_content[:150]}...")
except Exception as e:
    print(f"Verifique se o caminho do PDF esta correto: {e}")
````

````python
# --- 3. VECTOR DATABASE E EMBEDDINGS ---
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

try:
    embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
    vector_store = FAISS.from_documents(splits, embeddings)
    print("VectorDB FAISS carregado em RAM com sucesso!")

    # DEMONSTRACAO DE SIMILARIDADE: Garantindo que o banco funciona independentemente da IA.
    retriever_teste = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 2})
    teste_buscado = retriever_teste.invoke("produtos")
    
    print("\n[Demonstracao Retirada]")
    if len(teste_buscado) > 0:
        print(teste_buscado[0].page_content[:150])
except Exception as e:
    print(f"Verifique sua chave de OpenAI: {e}")
````

````python
# --- 4. RAG E PROMPT ENGINEERING ---
from langchain_openai import ChatOpenAI
from langchain_classic.chains import create_retrieval_chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.3)

system_prompt = (
    "Aja explicitamente como um assistente de clientes virtual amigavel e objetivo da QuantumFinance.\n"
    "Responda a duvida baseando-se unica e tao somente no contexto injetado abaixo.\n"
    "Jamais adicione informacoes que não estao no contexto.\n"
    "Se não houver a integracao no contexto corporativo para o que o usuario esta perguntando, recuse-se a responder, informando que seu dominio de ajuda e restrito a ferramentas e duvidas correspondentes sobre a QuantumFinance.\n"
    "Contexto provido de seguranca:\n"
    "{context}"
)

# Integracao das instrucoes Langchain
prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("human", "{input}"),
])

try:
    question_answer_chain = create_stuff_documents_chain(llm, prompt)
    retriever = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 3})
    rag_chain = create_retrieval_chain(retriever, question_answer_chain)

    demonstracao = rag_chain.invoke({"input": "Sou um cliente novo, como você pode me ajudar e o que exatamente á essa empresa?"})
    print("\n[RESPOSTA GERADA PELO RAG]")
    print(demonstracao["answer"])
except Exception as e:
    print(f"Verifique suas importacoes ou chave do endpoint: {e}")
````

````python
# ==============================================================
# PIPELINE FINAL CHATBOT CORPORATIVO - VALIDACAO DO PROFESSOR
# ==============================================================
import sys
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.vectorstores import FAISS
from langchain_classic.chains import create_retrieval_chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate

os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY # Coloquei essa chamado pq a API do chatgpt puxa as credenciais de var global ai forcei pegar do script

def central_chatbot(caminho_pdf):
    # Carregamento do Arquivo Corporativo e Fatiamento
    print("[Pipeline] Ingerindo base de conhecimento PDF e criando os fragmentos...")
    try:
        # Modulo langchain isolado da documentacao text_splitters (devido a nova formatacao da lib)
        docs = PyPDFLoader(caminho_pdf).load()
        splits = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150).split_documents(docs)
    except Exception as e:
        print(f"[Sistema] Nao foi possivel carregar o documento {caminho_pdf}. Erro: {e}")
        return

    # Processamento Vetorial - Embedds + FAISS In-memory
    print("[Pipeline] Executando embeddings de inteligencia artificial e armazenando no VectorDB (FAISS)...")
    try:
        vetores_bd = FAISS.from_documents(splits, OpenAIEmbeddings(model="text-embedding-3-small"))
        recuperador = vetores_bd.as_retriever(search_type="similarity", search_kwargs={"k": 4})
    except Exception as e:
         print(f"[Sistema] Falha nas chaves OpenAI ou conexao Vetorial: {e}")
         return
         
    # Integracao das restricoes System Prompt + LLM e Montagem da Cadeia 
    print("[Pipeline] Sincronizando LLM e regras via Prompt Engineering (Chat Prompt)...")
    try:
        llm_inteiro = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.2)
        regras = (
            "Assuma o papel de um Chatbot de atendimento receptivo chamado Assistente Virtual QuantumFinance.\n"
            "Sua missao obrigatoria e responder duvidas usando ESTREITAMENTE o contexto embutido.\n"
            "Se o solicitante abordar um tema que fuja do escopo extraido do manual da QuantumFinance, informe de modo muito respeitoso que voce so possui autorizacao e limite de conhecimento para abordar questoes referentes a QuantumFinance.\n"
            "Seja direto mas agradavel.\n"
            "--- CONTEXTO ---\n"
            "{context}"
        )
        
        prompt_completo = ChatPromptTemplate.from_messages([
            ("system", regras),
            ("human", "{input}"),
        ])
        
        rag_completo = create_retrieval_chain(recuperador, create_stuff_documents_chain(llm_inteiro, prompt_completo))
    except Exception as e:
        print(f"[Sistema] Falha durante a cadeia gerativa do RAG: {e}")
        return

    print("\n" + "="*55)
    print("      QUANTUMFINANCE - ATENDIMENTO DIGITAL ATIVO")
    print("      (digite 'sair', 'fechar' ou 'quit' para finalizar)   ")
    print("="*55)
    
    # Aqui, para ambiente simulado, iteramos um Chat
    while True:
        interacao = input("\n[Cliente] ")
        
        if not interacao.strip(): continue
        if interacao.strip().lower() in ['sair', 'fechar', 'quit']:
            print("\n[Assistente] Agradeco fortemente seu contato. Suporte QuantumFinance Finalizado.")
            break
            
        print("[Assistente] Buscando...")
        try:
           resposta = rag_completo.invoke({"input": interacao})
           print(resposta["answer"])
        except Exception as e:
           print(f"Desculpe, meu modulo generativo relata instabilidade. [{e}]")

# Iniciando ambiente
pdf_alvo = arquivo_pdf
try:
    central_chatbot(pdf_alvo)
except Exception as e:
    print(f'Erro critico durante a execucao master: {e}')
````
